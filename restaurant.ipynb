```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# Load the restaurant data
csv_file = os.path.join("resources", "restaurant_data.csv")

restaurant_df = pd.read_csv(csv_file, encoding = 'utf-8')
restaurant_df.head(10)
```

```python
# Check for missing data
restaurant_df.count()
```

```python
# Check data types
restaurant_df.dtypes
```

```python
# Create a country list for x-axis
country_list = ["American", "French", "Indian", "Italian", "Japanese", "Mexican"]
```

```python
# Group with "Cuisine" by each countries with average "Revenue"
group0 = restaurant_df.groupby(["Cuisine"])
group0 = group0[["Revenue"]].mean()
group0
```

```python
# Create a "Revenue" list 
revenue = group0["Revenue"].tolist()
revenue
```

```python
# Create a bar chart comparing "Revenue" by each countries
plt.bar(country_list, revenue, color = 'orange')

plt.title("Comparison between countries (revenue)")
plt.xlabel("Cuisine by each countries")
plt.ylabel("Revenue")

revenue_bar_path = os.path.join("results", "revenue_bar_comparison.png")
plt.savefig(revenue_bar_path)
```

```python
# Create a line graph comparing "Revenue" by each countries
plt.plot(country_list, revenue, color = 'green', label = "revenue")

plt.title("Comparison between countries (revenue)")
plt.xlabel("Cuisine by each countries")
plt.ylabel("Revenue")
plt.legend(loc = 'best')

revenue_line_path = os.path.join("results", "revenue_line_comparison.png")
plt.savefig(revenue_line_path)
```

```python
# Group with "Cuisine" by each countries with average "Weekend Reservations" and "Weekday Reservations"
group3 = restaurant_df.groupby(["Cuisine"])
group3 = group3[["Weekend Reservations", "Weekday Reservations"]].mean()
group3
```

```python
# Create a "Weekend Reservations" list
weekend = group3["Weekend Reservations"].tolist()
weekend
```

```python
# Create a "Weekday Reservations" list
weekday = group3["Weekday Reservations"].tolist()
weekday
```

```python
# Create a line graph comparing "Weekend Reservations" and "Weekday Reservations" by each countries
plt.plot(country_list, weekend, color = 'red', label = 'weekday reservation')
plt.plot(country_list, weekday, color = 'blue', label = 'weekend reservation')

plt.title("Comparison between countries (Reservations)")
plt.xlabel("Cuisine by each countries")
plt.ylabel("Reservations")
plt.legend(loc = 'best')

weekend_weekday_path = os.path.join("results", "weekend_weekday_line_comparison.png")
plt.savefig(weekend_weekday_path)
```
